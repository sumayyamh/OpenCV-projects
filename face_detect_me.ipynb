```python
import cv2
img=cv2.imread(r'D:\Computer Vision\OpenCv\face detection\Many People.jpg')
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(con_img)
```

```python
face_clf=cv2.CascadeClassifier(r'D:\Computer Vision\OpenCv\face detection\haarcascade_frontalface_default.xml')
face_clf
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

```python
face=face_clf.detectMultiScale(img_gray,scaleFactor=1.1,minNeighbors=10)
face
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),color=(0,255,0),thickness=3)
plt.imshow(con_img)
```

```python
video=cv2.VideoCapture(0)
```

```python
video=cv2.VideoCapture(0)
while True:
    success,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_clf.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
    print(len(faces))
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),thickness=3)
    cv2.imshow('Video',frame)
    if cv2.waitKey(1) & 0XFF==27:
        break
video.release()
cv2.destroyAllWindows()
```
