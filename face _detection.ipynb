```python
import cv2
img=cv2.imread(r'D:\Computer Vision\OpenCv\face detection\monster cast.jpg')
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(converted_img)
```

```python
face_detect_classifier=cv2.CascadeClassifier(r'D:\Computer Vision\OpenCv\face detection\haarcascade_frontalface_default.xml')
```

```python
face_detect_classifier
```

```python
#gray scale image have better results.
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
face=face_detect_classifier.detectMultiScale(img_gray,scaleFactor=1.1,minNeighbors=5)
face
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(converted_img,(x,y),(x+w,y+h),color=(0,255,0),thickness=4)
plt.imshow(converted_img)
```

```python
video=cv2.VideoCapture(r'D:\Computer Vision\OpenCv\face detection\video.mp4')
while True:
    success,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_detect_classifier.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
    print(len(faces))
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),thickness=3)
        cv2.imshow('Video',frame)
        if cv2.waitKey(1) & 0XFF==ord('k'):
                 break
video.release()
cv2.destroyAllWindows()   
        
```
